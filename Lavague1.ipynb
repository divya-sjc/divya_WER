```python
# draft flow

# creation flow amazon example
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
from datetime import datetime, timedelta
import random
import json
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException, WebDriverException
from selenium.common.exceptions import UnexpectedAlertPresentException
from selenium.webdriver import ActionChains
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.actions.action_builder import ActionBuilder
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities


driver_path = '/Users/divyac/Documents/Required files/chromedriver-mac-arm64/chromedriver'
driver_service = Service(driver_path)
options = webdriver.ChromeOptions()
options.add_argument("--disable-blink-features=AutomationControlled")
# options.add_argument("--headless")  # Run headless Chrome (no GUI)
options.add_argument("--disable-gpu")
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)
options.add_argument("--guest")
# driver = webdriver.Chrome(options=options)
options.set_capability("goog:loggingPrefs", {"performance": "ALL"})
    
# Setup ChromeDriver
driver = webdriver.Chrome(service=driver_service, options=options)

main_window_handle = driver.current_window_handle
n = str(random.randint(1, 100))


def xpath_click(str1, time1):    #waiting_for_element
    try:
        time.sleep(1)
        element = WebDriverWait(driver, time1).until(EC.visibility_of_element_located((By.XPATH, str1)))
        element.click()
    except:
        print(f"Error: Element with XPath '{str1}' not found within the timeout period.")
        network_logs()

def id_click(str1, time1):     #waiting_for_element_id
    try:
        time.sleep(1)
        element = WebDriverWait(driver, time1).until(EC.visibility_of_element_located((By.ID, str1)))
        element.click()
    except:
        print(f"Error: Element with ID '{str1}' not found within the timeout period.")
        network_logs()

def xpath_send(str1, time1, value):    #waiting_for_element
    try:
        time.sleep(1)
        element = WebDriverWait(driver, time1).until(EC.visibility_of_element_located((By.XPATH, str1)))
        element.clear()
        element.send_keys(value)
    except:
        print(f"Error: Element with XPath '{str1}' not found within the timeout period.")
        network_logs()

def id_send(str1, time1, value):     #waiting_for_element_id
    try:
        time.sleep(1)
        element = WebDriverWait(driver, time1).until(EC.visibility_of_element_located((By.ID, str1)))
        element.clear()
        element.send_keys(value)
    except:
        print(f"Error: Element with ID '{str1}' not found within the timeout period.")
        network_logs()

def scroll(driver, direction):
    if direction == 'down':
        driver.execute_script("window.scrollBy(0, 1000);")
    elif direction == 'up':
        driver.execute_script("window.scrollTo(0, 0);")
    elif direction == 'left':
        driver.execute_script("window.scrollBy(-1000, 0);")
    elif direction == 'right':
        driver.execute_script("window.scrollBy(1000, 0);")


def add_custom_capability(str1, add_para, parameter_name, parameter_description, del1):
    time.sleep(2)
    id_click('add-new-cap-btn', 20)   # Add capability
    time.sleep(2)
    id_click("create-custom-capability-btn", 20) #custom capability button
    time.sleep(2)
    # click on the edit description of the capability text box
    try:
        element = WebDriverWait(driver, 25).until(EC.visibility_of_element_located((By.CSS_SELECTOR,  'div.monaco-editor')))
        editor_container = driver.find_element(By.CSS_SELECTOR,  'div.monaco-editor')
        act = ActionChains(driver)
        act.click(editor_container).perform()
        driver.implicitly_wait(500)
        try:
            driver.implicitly_wait(500)
            script_get_value = 'return monaco.editor.getModels()[0].getValue();'
            editor_text = driver.execute_script(script_get_value)
            replace_string = str1
            script_set_value = f'monaco.editor.getModels()[0].setValue(`{replace_string}`);'
            driver.execute_script(script_set_value)
        except:
            print("unable to edit")
            network_logs()
            return
        xpath_click("//button[normalize-space()='Generate Capability']", 30)
        if (add_para == 1):
            time.sleep(5)
            add_parameter(parameter_name, parameter_description, del1)
        #adding group
        id_click('groups-plusicon-btn', 20)
        id_click('new-group-select', 20)
        ele = WebDriverWait(driver, 20).until(EC.visibility_of_element_located((By.ID, 'add-new-group-input')))
        ele.send_keys('shopping')
        ele.send_keys(Keys.RETURN)
        id_click('add-new-group-btn', 20)
        id_click("edit-cap-config-save-btn", 25)
        driver.implicitly_wait(200)
        return
    except:
        xpath_click("//button[normalize-space()='Cancel']", 20)
        print("cancelled")
        network_logs()
        return

def edit_assistant_add_custom_capability(str1, add_para, parameter_name, parameter_description):
    time.sleep(2)
    id_click('add-new-cap-btn', 20)   # Add capability
    time.sleep(2)
    id_click("create-custom-capability-btn", 20) #custom capability button
    time.sleep(2)
    # click on the edit description of the capability text box
    try:
        element = WebDriverWait(driver, 25).until(EC.visibility_of_element_located((By.CSS_SELECTOR,  'div.monaco-editor')))
        editor_container = driver.find_element(By.CSS_SELECTOR,  'div.monaco-editor')
        act = ActionChains(driver)
        act.click(editor_container).perform()
        driver.implicitly_wait(500)
        try:
            driver.implicitly_wait(500)
            script_get_value = 'return monaco.editor.getModels()[0].getValue();'
            editor_text = driver.execute_script(script_get_value)
            replace_string = str1
            script_set_value = f'monaco.editor.getModels()[0].setValue(`{replace_string}`);'
            driver.execute_script(script_set_value)
            xpath_click("//button[normalize-space()='Generate Capability']", 30)
            #adding group
            id_click('groups-plusicon-btn', 20)
            id_click('new-group-select', 20)
            ele = WebDriverWait(driver, 20).until(EC.visibility_of_element_located((By.ID, 'add-new-group-input')))
            ele.send_keys('casual')
            ele.send_keys(Keys.RETURN)
            id_click('add-new-group-btn', 20)
            id_click("edit-capability-compile-btn", 10000)     #compile
            WebDriverWait(driver, 7000).until(EC.element_to_be_clickable((By.ID, 'edit-capability-done-btn')))
            id_click('edit-capability-done-btn', 50)       #done
            time.sleep(1)
            return
        except:
            print("unable to edit")
            network_logs()
            return
    except:
        xpath_click("//button[normalize-space()='Cancel']", 20)
        print("cancelled")
        network_logs()
        return


def add_parameter(parameter_name, parameter_description, del1):
    try:
        id_click("specification-plusicon-btn", 25)
        try:
            id_send("parameter-name-0", 25, parameter_name)                    
            id_click("parameter-datatype-0", 25)
            id_click("parameter-datatype-0-datatype-1", 20)
            id_send("parameter-description-0", 25, parameter_description)
            id_click("parameter-required-0", 10)
            time.sleep(1)
            # print("parameter ", parameter_name, " is added")
            # Removing one parameter
            if(del1 ==1):
                time.sleep(2)
                try:
                    id_click("parameter-delete-0", 10)
                    time.sleep(2)
                    # print("Deleting a parameter works")
                    time.sleep(5)
                    add_parameter('category', 'category of the product being searched', 0)
                except:
                    print("could not delete a parameter")
                    network_logs()
        except:
            id_click("parameter-delete-0",10)
            driver.implicitly_wait(200)
            print("Deleting a parameter as it was not properly created")
            network_logs()
    except:
        print("Parameter could not be added")      
        network_logs()
    input_element = driver.find_element(By.ID, "parameter-name-5")
    if(input_element.get_attribute('value') == 'filters'):
        id_click("parameter-datatype-5", 25)
        id_click("parameter-datatype-0-datatype-5", 20)
            

def login_cred(user, password):
    try:
        time.sleep(2)
        id_send("email-login", 20, user)
        id_send("password-login", 5, password)
        id_click("login-btn", 5)
        return 1
    except:
        network_logs()
        return 0

def register_user(user, password):
    try:
        time.sleep(2)
        id_send("email-signup", 20, user)
        id_send("password-signup", 5, password)
        id_click("signup-btn", 5)
        return 1
    except:
        network_logs()
        return 0


def log_out():
    try:
        time.sleep(1)
        prof_pic = WebDriverWait(driver, 15).until(EC.visibility_of_element_located((By.ID, "profile-image")))
        driver.execute_script("arguments[0].scrollIntoView(true);", prof_pic)
        actions = ActionChains(driver)
        actions.move_to_element(prof_pic).click().perform()
        time.sleep(1)
        id_click("logout-icon", 5)
        time.sleep(2)
        return 1
    except:
        network_logs()
        return 0

def create_org():
    try:
        time.sleep(2)
        org = "sample"+n+" organisation"
        id_send("create_org_textfield", 20, org)
        id_click("create-org-btn", 5)
        # print("Creating an org works as expected")
        return 1
    except:
        network_logs()
        return 0

def close_coachmark():
    try:
        id_click("check-icon", 10)
        return 1
    except:
        network_logs()
        return 0
    
def google_register(user, password):
    try:
        time.sleep(2)
        try:
            xpath_send("/html/body/div[1]/div[1]/div[2]/c-wiz/div/div[2]/div/div/div[1]/form/span/section/div/div/div[1]/div/div[1]/div/div[1]/input", 25, user)      #sign in with gmail account - email
            xpath_click('//*[@id="identifierNext"]/div/button/span', 20)
        except:
            network_log()
        try:
            xpath_send('//*[@id="password"]/div[1]/div/div[1]/input', 20, password)      #sign in with gmail account - password
            xpath_click('//*[@id="passwordNext"]/div/button/span', 30)
        except:
            network_log()
    except:
        network_logs()
        time.sleep(2)

def switch_window(wind:str):
    if(wind == "new"):
        try:
            WebDriverWait(driver, 30).until(EC.new_window_is_opened(driver.window_handles))
            # print("New window opened.")
            window_handles = driver.window_handles
            # print(f"All window handles: {window_handles}")
            new_window_handle = None
            for handle in window_handles:
                if handle != main_window_handle:
                    new_window_handle = handle
                    break
            if not new_window_handle:
                raise Exception("New window handle not found.")
            else:
                driver.switch_to.window(new_window_handle)
        except:
            network_logs()
            return
    elif(wind == 'main'):
        try:
            driver.switch_to.window(main_window_handle)
            return
        except:
            network_logs()
            return
    return

def pg_textbox_test(t:str):
    id_send('playground-input', 20, t)
    id_click('playground-send-btn', 5) #enter button
    time.sleep(7)
    try:
        ele = driver.find_element(By.XPATH, "//div[@class='relative']//pre/code")
        sspan = ele.find_elements(By.CSS_SELECTOR, 'span.token')
        for x in sspan:
            print(x.text, end = "")
    except:
        network_logs()


def network_logs():
    logs = driver.get_log('performance')
    error_logs = []
    for log in logs:
        log_message = json.loads(log['message'])['message']
        if 'Network.responseReceived' in log_message['method']:
            response = log_message['params']['response']
            status = response.get('status')
            if status >= 400:  # Check for client and server errors
                error_logs.append(log_message)
        
        # Print out the error network logs for inspection
    for log in error_logs:
        print(json.dumps(log, indent=2))

def cap_smart_knowledge():
    try:
        id_click("knowledge-group-plus-btn", 20)
        id_click("create-knowledge-btn", 20)  # create knowledge
        id_send("knowledge-group-name-input", 20, "Return policy") #knowlwdge name
        id_send("knowledge-group-desc-input", 10, "Amamzon's return policy details") #knowledge description
        id_click("add-html-urls-btn", 20)
        id_send("add-knowledge-url-input", 20, "https://www.amazon.com/")
        id_click("knowledge-url-input-plusicon", 10)
        time.sleep(3)
        try:
            id_click("add-html-knowledge-submit-btn", 20)
            id_click("knowledge-group-create-btn", 20)    #create button
            id_click("import-knowledge-btn-0", 20)   #import
            # WebDriverWait(driver, 500).until(EC.visibility_of_element_located((By.XPATH, "/html/body/div[2]/div[2]/div/div/div[2]/div/div[1]/div/div/div/div[3]/div[3]/div/div/div[1]/div/div/div/svg")))
            return 1
        except:
            network_logs()
            return 0
    except:
        network_logs()
        return 0
        

def cap_static_knowledge():
    try:
        id_click("knowledge-group-plus-btn", 20)
        id_click("import-knowledge-btn-1", 20)   #import from store (1st)
        # WebDriverWait(driver, 500).until(EC.visibility_of_element_located((By.XPATH, "/html/body/div[2]/div[2]/div/div/div[2]/div/div[1]/div/div/div/div[3]/div[3]/div/div/div[1]/div/div/div/svg")))
        
        id_click("knowledge-group-plus-btn", 20)
        id_click("create-knowledge-btn", 20)  # create knowledge
        id_send("knowledge-group-name-input", 20, "Return policy") #knowlwdge name
        id_send("knowledge-group-desc-input", 10, "Amazon's return policy details") #knowledge description
        id_click("knowledge-type-select", 10)   # choose the type of knowledge
        id_click("knowledge-type-select-datatype-1", 20)  #static
        id_click("knowledge-url-radio", 20)   #url radio button
        id_send("knowledge-url-input", 20, "https://www.amazon.in/gp/help/customer/display.html?nodeId=202111910")
        id_click("knowledge-group-create-btn", 20)    #create button
        # webDriverWait(driver, 500).until(EC.visibility-of_element_located(By.XPATH, ""))
        id_click("import-knowledge-btn-0", 20)  #import newly created knowledge
        id_click("ground-to-parameter-checkbox", 20) #ground to a parameter
        id_click("ground-to-parameter-select", 20)
        id_click("ground-to-parameter-select-datatype-2", 10)
        
        id_click("knowledge-group-plus-btn", 20)
        id_click("create-knowledge-btn", 20)  # create knowledge
        id_send("knowledge-group-name-input", 20, "Refund policy") #knowlwdge name
        id_send("knowledge-group-desc-input", 10, "Amamzon's refund policy details") #knowledge description
        id_click("knowledge-type-select", 10)      # choose the type of knowledge
        id_click("knowledge-type-select-datatype-1", 20) #static
        id_click("knowledge-text-radio", 20)   #textarea radio button
        description = "This may help. What's Amazon's refund policy? What's Amazon's refund policy? When returning an item, you have the option to choose your preferred refund method in the Online Returns Center. Your refund amount and method may vary. Check the payment method refunded and the status of your refund in Your Orders."
        id_click("knowledge-text-input", 20)
        id_send("knowledge-text-input", 20, description)
        time.sleep(2)
        id_click("knowledge-group-create-btn", 20)    #create button
        time.sleep(2)
        id_click("import-knowledge-btn-0", 20)  #import newly created knowledge
        time.sleep(2)
    except:
        network_logs()

try:
    driver.execute_cdp_cmd('Network.enable', {})        # open network tab
    
    driver.get("https://studio.conva.ai/login")    #Prod
    # driver.get("https://magic-studio-stage.delightfulforest-40b0615c.centralindia.azurecontainerapps.io/login")   #stage
    
    # Login with correct user credentials
    a = 0
    try:
        user = "divya@slanglabs.in"
        x = login_cred(user, "abc@123")
        time.sleep(2)
        try:
            element = WebDriverWait(driver, 40).until(EC.visibility_of_element_located((By.ID, 'create-assistant-btn')))
            print(f"Test case 11: Login with valid credentials work {user} - Pass")
            a = 1
        except:
            print(f"Test case 11: Registering an existing user credentials work {user}- Fail")
            a = 0
            network_logs()
    except:
        print(f"Unable to access email and password")
        a = 0
        network_logs()
    

    #create Assistant -  grocker
    if(a == 1):
        b = 0
        try:
            id_click("create-assistant-btn", 20)    #clicking create assistant btn
            id_send("app-url-input", 20, "https://www.amazon.com/")
            time.sleep(1)
            id_click("regenerate-app-details-btn", 20)
            WebDriverWait(driver, 40).until(EC.element_to_be_clickable((By.ID, "add-capabilities-create-btn")))
            #change the app name
            id_send("app-name-input", 40, "Automated_Amazon_draft")
            time.sleep(1)
            description_value = "Free shipping on millions of items. Get the best of Shopping and Entertainment with Prime. Enjoy low prices and great deals on the largest selection of everyday essentials and other products, including grocery, fashion, home, beauty, electronics, Alexa Devices, sporting goods, toys, automotive, pets, baby, books, video games, musical instruments, office supplies, house hold, fitness, gift, personal care, pharmacy, furniture and more."
            id_send('app-desc-input', 10, description_value)
            #change the app category
            category_value = "grocery, fashion, home, beauty, electronics, Alexa Devices, sporting goods, toys, automotive, pets, baby, books, video games, musical instruments, office supplies, house hold, fitness, personal care, pharmacy, furniture, gift,  fitness, mother care and more."
            id_send("categories-input", 10, category_value)
            try:
                id_click("add-capabilities-create-btn", 10)
                print(f"Test case 12: App grokker works - Pass")
                b = 1
            except:
                time.sleep(3)
                print(f"Test case 12: App grokker doesn't work - Fail")
                b = 0
                network_logs()
        except:
            print(f"Unable to get response from grokking / edit the grokker form")
            b = 0
            network_logs()
   
    
    #Adding capability - Navigation
    if(b == 1):
        try:
            id_click('add-new-cap-btn', 20)        #Add capability btn
            id_click('import-btn-2', 35)      #navigation import
            time.sleep(1)
            x = cap_smart_knowledge()
            if (x ==1):
                id_click("edit-cap-config-save-btn", 40)    # save btn
                print(f"Test case 13: Importing capability from store (Navigation) - Pass")
            else:
                network_log()
        except:
            print(f"Test case 13: Unable to importing capability from store (Navigation) - Fail")
            network_logs()

    
    # Adding Capability
    # Add custom capability for Assisted Shopping
    if(b==1):
        c = 0
        try:
            replace_string = """Ability to enhance the user in shopping experience through an "assisted shopping" capability to find products quickly and accurately.. 
you MUST provide an appropriate response has "message"
You SHOULD identify the search term from the input query.  Note that the search term must be a product name with or without the brand name. It MUST NOT have conjunctions in it. 
If the user gives more than one product in the input query from any category supported by the app, then you MUST consider this has shopping list. Add the first product to the search term and the remaining products to "related_products". Do NOT add this shopping list to the cart until the user wants you to add to cart
You **MUST** provide 4 to 6 products related to the search_term has "related_products". You must NOT provide related products has product 1, product 2, product 3 etc. or product A, product B, product C etc. They all must be valid product names.
Identify the brand name only if the user is looking for a particular brand.
Depending on the users need to add the product to the cart or not, the cart value must be 0 or 1. 
If the user gives the quantity and measurement unit, it must be stored in respective parameter.
If the user asks for any recommendation or suggestion questions like gift ideas, you MUST suggest some product that align with the input query and the search term must be a product.
Enhance user to search by array of dictionary filters such as: gender, age, size, color, price, and rating to narrow down your search.
You MUST not answer any questions that do not adhere to the app category.
"""
            add_custom_capability(replace_string, 1, 'Filter', 'filter by age, gender, size etc', 1)
            print(f"Test case 14: Adding custom capability and editing parameters - Pass")
            c = 1
        except:
            print(f"Test case 14: Unable to add custom capability and edit parameters - Fail")
            c = 0
            network_logs()

    #close the 
    if(c==1):
        d=0
        try:
            xpath_click("/html/body/div[2]/div/div/div[1]/button", 20)
            id_click("assistant-wizard-cancel-btn", 20)
            xpath_click("/html/body/div[3]/div/div/div[1]/button", 20)
            id_click("assistant-wizard-proceed-btn", 20)
            d=1
        except:
            network_logs()



    #opening draft assistant
    driver.refresh()
    e=0
    if((d==1) or (c==1)):
        try:
            xpath_click("//p[normalize-space()='Automated_Amazon_draft']", 20)
        
            # Adding capability - Domain FAQ
            try:
                id_click("add-new-cap-btn", 20)
                id_click("import-btn-0", 20)
                cap_static_knowledge()
                id_click("edit-cap-config-save-btn", 25)
            except:
                network_logs()

            #build assistant
            try:
                id_click("add-capabilities-review-btn", 40)
                id_click("build-assistant-btn", 25)
                try:
                    WebDriverWait(driver, 10000).until(EC.visibility_of_element_located((By.ID, "assistant-ready")))
                    id_click("try-assistant-btn", 10)
                    driver.implicitly_wait(2.5)
                    print(f"Test case 15: Compiling, setting up and configuring - Pass")
                    time.sleep(1)
                    ele = driver.find_element(By.ID, 'assistant-id-input')
                    print("Asistid: ", ele.get_attribute("value"))
                    ele = driver.find_element(By.ID, 'api-key-input')
                    print("API Key: ", ele.get_attribute("value"))
                    ele = driver.find_element(By.ID, 'assistant-version-input')
                    print("Version: ", ele.get_attribute("value"))
                    e=1
                except:
                    print(f"Test case 15: Unable to compiling, setting up and configuring - Fail")
                    d=0
                    network_logs()
            except:
                print("Unable to click on review and build capability btn")
                network_logs()
        except:
            network_logs()

    
    # Playground testing
    if(e==1):
        f=0
        try:
            id_click('playground-btn', 20) #Click on playgroup tab
            time.sleep(3)
            f=1
        except:
            network_logs()

    # First hint is clicked
    if(f==1):
        try:
            id_click('example-ques-0', 30) # click on first hint
            id_click("playground-expand-btn", 35) # click on expand json
            # id_click('playground-code-undefined', 20) # access  
            ele = driver.find_element(By.XPATH, "//div[@class='relative']//pre/code")
            time.sleep(2)
            sspan = ele.find_elements(By.CSS_SELECTOR, 'span.token')
            for x in sspan:
                print(x.text, end = "")
            time.sleep(2)
            print()
            print(f"Test case 16: Playground Testing - First hint - Pass")
            for i in range(0, 2):
                print("")
        except:
            print(f"Test case 16: Playground Testing - First hint - Fail")
            network_logs()

    # input box testing
    if(f==1):
        try:
            id_send('playground-input', 20, 'what can i wear to match my red bell sleeves tops?')
            time.sleep(1)
            id_click('playground-send-btn', 5) #enter button
            time.sleep(7)
            ele = driver.find_element(By.XPATH, "/html/body/div/div/div/div[2]/main/div/div/div/div/div[2]/div/div/div[1]/div[1]/div[2]/div[2]/div/div[2]/div/pre/code")
            sspan = ele.find_elements(By.CSS_SELECTOR, 'span.token')
            for x in sspan:
                print(x.text, end = "")
            time.sleep(2)
            print()
            print(f"Test case 17: Playground Testing - Text input with conversation history enabled- Pass")
            for i in range(0, 2):
                print("")
        except:
            print(f"Test case 17: Playground Testing - Text input with conversation history enabled - Fail")
            network_logs()

    # removing the conversation history
    if(f==1):
        try:
            id_click("conv-history-checkbox", 20)
            pg_textbox_test("i need india gate basmati rice")
            print(f"Test case 18: Playground Testing - Conversation history disabled - Pass")
            for i in range(0, 2):
                print("")
        except:
            print(f"Test case 18: Playground Testing - Conversation history disabled - Fail")
            network_logs()
    
    # selecting group
    if(f==1):
        pg_textbox_test("who are you?")
        try:
            capability_option = WebDriverWait(driver, 3).until(EC.element_to_be_clickable(By.ID, 'playground-cap-select'))
            print(f"Test case 19: Playground Testing - capability selection is diabled when group is selected - Fail")
            network_logs()
        except:
            print(f"Test case 19: Playground Testing - capability selection is diabled when group is selected - Pass")
            for i in range(0, 2):
                print("")
    
    # selecting capability
    if(f==1):
        id_click("capability-radio", 20)  # capability
        id_click("playground-cap-select", 20)   # capability option
        try:
            id_click("playground-cap-select-datatype-3", 20)     #assisted shopping selection
        except:
            network_log()
        pg_textbox_test("show me sont tv 4K")
        try:
            capability_option = WebDriverWait(driver, 3).until(EC.element_to_be_clickable(By.ID, 'playground-groups-select'))
            print(f"Test case 20: Playground Testing - Group selection is diabled when capability is selected - Fail")
            network_logs()
        except:
            print(f"Test case 20: Playground Testing - Group selection is diabled when capability is selected - Pass")
            for i in range(0, 2):
                print("")


     # Logs - display all values
    flag = 1
    try:
        id_click("monitoring-btn", 20)         #Log monitor is clicked
        today = datetime.today().date()
        tomorrow = today + timedelta(days=1)
        yesterday = today - timedelta(days=1)
        tomorrow_str = tomorrow.strftime('%d/%m/%y')
        yesterday_str = yesterday.strftime('%d/%m/%y')
        print("Yesterday:", yesterday_str)
        print("Tomorrow:", tomorrow_str)
        time.sleep(3)
        element = WebDriverWait(driver, 20).until(EC.visibility_of_element_located((By.ID, 'end-date-input')))  
        # element.click()
        element.clear()
        element.send_keys(tomorrow_str)
        element = WebDriverWait(driver, 20).until(EC.visibility_of_element_located((By.ID, 'start-date-input')))  
        # element.click()
        element.clear()
        element.send_keys(yesterday_str)
        id_click("playground-refetch-btn", 20)
        try:
            time.sleep(4)
            WebDriverWait(driver, 100).until(EC.visibility_of_element_located((By.ID, 'log-viewer-row-0')))
            try:
                table = driver.find_element(By.TAG_NAME, "tbody")
                try:
                    rows = table.find_elements(By.TAG_NAME, 'tr')
                    for row in rows:
                        try:
                            cells = row.find_elements(By.TAG_NAME, 'td')
                            for cell in cells:
                                print(cell.text, end=' ')
                            print(f"Test case 21: Log monitor - viewing the entire table works - Pass")
                            print()
                        except:
                            print("td not found")
                            print(f"Test case 21: Log monitor - viewing the entire table doesn't works - Fail")
                            flag = 0
                except:
                    print("tr not found")
                    print(f"Test case 21: Log monitor - viewing the entire table doesn't works - Fail")
                    flag = 0
            except:
                print("table not found")
                print(f"Test case 21: Log monitor - viewing the entire table doesn't works - Fail")
                flag = 0
                print()
        except:
            network_logs()
    except:
        network_logs()


    # Logs - 1 value details
    if (flag == 1):
        try:
            id_click('log-viewer-row-0', 20)
            time.sleep(3)
            output = driver.find_element(By.ID, "side-panel-children")
            divs = output.find_elements(By.CLASS_NAME, 'space-y-1')
            for row in divs:
                cells = row.find_elements(By.TAG_NAME, 'p')
                for cell in cells:
                    print(cell.text, end=' ')
                print()
            print(f"Test case 22: Log monitor - viewing one row's complete detail works- Pass")
            for i in range(0, 2):
                print("")
        except:
            print(f"Test case 22: Log monitor - viewing one row's complete detail doesn't work- Fail")
            network_logs()
            
        
finally:
    driver.quit()
    
```
